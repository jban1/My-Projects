```python
import random
from collections import Counter
from operator import itemgetter

suit = ('Hearts','Diamonds', 'Clubs', 'Spades')
face = ('Ace', 'Deuce', 'Three', 'Four','Five','Six','Seven','Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King')

#rank and face - same

face_seq = []
j = 1
for i in face:
    face_seq = face_seq + [(i, j)]
    j +=1

#print (face_seq)

deck=[]
for i in suit:
    for j in face:
        deck = deck +[(j,i)]
        
random.shuffle(deck)

#deck = [('Ace', 'Hearts'), ('Five', 'Clubs'), ('Four', 'Hearts'), ('Deuce', 'Hearts'), ('Three', 'Hearts')] 
for a, b in deck:
    c = (a + ' of ' + b)
    #print (f'{c:<20}', end = '              ')
    
#print("\n")

pick = []
for i in range(0,5):
    pick = pick + [deck[i]]


print ('For your turn on Poker - you get the first 5 cards', '\n')
    
for a, b in pick:
    c = (a + ' of ' + b)
    print (f'{c:<20}', end = '              ')
    
print("\n")


pickws = []
for i1, i2 in pick:
    for j1, j2 in face_seq:
        if (i1==j1):
            pickws = pickws + [(j2, i1, i2)]
            #print (i,j)
pickws.sort()


storesuit = []
storerank = []

c1 = 0

for a, b, c in pickws:
    
    storerank = storerank + [a]
    
    if c not in storesuit:
        storesuit = storesuit + [c]
    
for i in range(0,4):
    if (storerank[i+1]-storerank[i] == 1):
        c1 += 1

c2 = Counter(storerank)
#print (c2, "\n")        

highrank = []
for i, j in sorted(c2.items()):
    #print (i,j)
    highrank += [j]

    
highrank.sort(reverse=True)
#print ("highrank", highrank)

rank = []
for i in range (0,len(pickws)):
    if i not in rank:
        rank = rank + [pickws[i][1]]

#print (rank)

c3 = Counter(rank)
c4 = sorted(c3.items(), key=itemgetter(1), reverse=True)
#print (c4)


        
print ('RESULT', '\n')

if (len(storesuit) == 1) and (c1 == 4):
    print ('1. Straight Flush - contains five cards of sequential rank, all of the same suit', '\n')
    print ('you have '  )

elif (len(storesuit) == 1) and (c1 != 4):
    print ('4. Flush - five cards all of the same suit, not all of sequential rank')

elif (len(storesuit) != 1) and (c1 == 4):
    print ('5. Straight - contains five cards of sequential rank, not all of the same suit') 

elif (highrank[0] == 4):
    print ('2. Four of a kind - contains four cards of one rank and one card of another rank')
    print ('   You have 4', c4[0][0], 'and 1 card of other rank')
    
elif (highrank[0] == 3) and (highrank[1] == 2):
    print ('3. Full house - contains three cards of one rank and two cards of another rank')
    print ('   You have 3', c4[0][0], 'and 2 cards of other ranks')
    
elif (highrank[0] == 3) and (highrank[1] == 1):
    print ('6. Three of a kind - contains three cards of one rank and two cards of two other ranks')
    
elif (highrank[0] == 2) and (highrank[1] == 2):
    print ('7. Two pair - contains two cards of one rank, two cards of another rank and one card of a third rank')
    print ('   You have 2', c4[0][0], ' and 2', c4[1][0],'and 1 card of other rank')
    
elif (highrank[0] == 2) and (highrank[1] == 1):
    print ('8. One pair - contains two cards of one rank and three cards of three other ranks', '\n')
    print ('   You have 2', c4[0][0], 'and 1 card of other ranks')
    
else:
    print ('9. High Card - nothing', '\n')
           
```
